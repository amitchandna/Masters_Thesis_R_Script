```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Nitrogen_Carbon_Values.csv')
```

```python
Xs = df.copy()
Xs = Xs.drop(columns='Treatment')
```

```python
X = Xs
y = df['Treatment']
```

```python
X = X.drop(columns=['Sample', 'Nitrogen (%)','Carbon(%)','Percentage of Phos'])
X
```

```python
model = LinearRegression()
target = y 
```

```python
model.fit(X,y)
model.score(X,y)
sns.pairplot(df, hue='Treatment')
plt.savefig('Pairplot.tiff')
```
